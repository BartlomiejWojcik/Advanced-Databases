```python
import pandas as pd
```

```python
data = pd.read_csv('superbowl.csv') 
data.head()
```

```python
print(data.columns)
```

```python
data.info()
```

```python
state_list = pd.DataFrame(data['State'].unique(), columns=['state'])
state_list.index.name = 'id'
state_list.head()
```

```python
city_list = data[['City','State']].drop_duplicates().reset_index().drop(columns = ['index']);
city_list.index.name = 'id'

city_list = city_list.rename(columns = {'State':'state_id'})
city_list = city_list.rename(columns = {'City':'city'})

city_list['state_id'] = city_list['state_id'].map(lambda x:  state_list[state_list['state'] == x].index.values.astype(int)[0])

city_list.head()
```

```python
stadium_list = data[['Stadium','City']].drop_duplicates().reset_index().drop(columns = ['index']);
stadium_list.index.name = 'id'

stadium_list = stadium_list.rename(columns = {'City':'city_id'})
stadium_list = stadium_list.rename(columns = {'Stadium':'stadium'})

stadium_list['city_id'] = stadium_list['city_id'].map(lambda x:  city_list[city_list['city'] == x].index.values.astype(int)[0])

stadium_list.head()
```

```python
player_list = pd.DataFrame(data['MVP'].unique(), columns=['player'])
player_list.index.name = 'id'
player_list.head()
```

```python
team_winner_list = pd.DataFrame(data['Winner'].unique(), columns=['team'])
team_winner_list.index.name = 'id'
team_winner_list.head()
```

```python
team_loser_list = pd.DataFrame(data['Loser'].unique(), columns=['team'])
team_loser_list.index.name = 'id'
team_loser_list.head()
```

```python
team_list = team_winner_list.append(team_loser_list, ignore_index = True)
team_list = team_list.drop_duplicates("team").reset_index().drop(columns = ['index'])
team_list.index.name = 'id'
team_list.head()
```

```python
superbowl_list = data.drop_duplicates().drop(columns = ['City','State']);

superbowl_list = superbowl_list.rename(columns = {'Date':'date'})
superbowl_list = superbowl_list.rename(columns = {'SB':'sb'})
superbowl_list = superbowl_list.rename(columns = {'Winner':'winner_id'})
superbowl_list = superbowl_list.rename(columns = {'Winner Pts':'winner_points'})
superbowl_list = superbowl_list.rename(columns = {'Loser':'loser_id'})
superbowl_list = superbowl_list.rename(columns = {'Loser Pts':'loser_points'})
superbowl_list = superbowl_list.rename(columns = {'MVP':'mvp_id'})
superbowl_list = superbowl_list.rename(columns = {'Stadium':'stadium_id'})

superbowl_list['winner_id'] = superbowl_list['winner_id'].map(lambda x:  team_list[team_list['team'] == x].index.values.astype(int)[0])
superbowl_list['loser_id'] = superbowl_list['loser_id'].map(lambda x:  team_list[team_list['team'] == x].index.values.astype(int)[0])
superbowl_list['mvp_id'] = superbowl_list['mvp_id'].map(lambda x:  player_list[player_list['player'] == x].index.values.astype(int)[0])
superbowl_list['stadium_id'] = superbowl_list['stadium_id'].map(lambda x:  stadium_list[stadium_list['stadium'] == x].index.values.astype(int)[0])
superbowl_list.index.name = 'id'
superbowl_list.head()
```

```python
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
```

```python
db_string = "postgresql://postgres:postgres@localhost:5432/superbowl"
engine = create_engine(db_string)
Base = declarative_base()
```

```python
from sqlalchemy import Column, Integer, String, Float, ForeignKey, Sequence, CheckConstraint, UniqueConstraint
```

```python
class Player(Base):
    __tablename__ = 'players'
    __table_args__ = (
        UniqueConstraint('player'),
    )
    id = Column(Integer, primary_key=True)
    player = Column(String(50))
    
class State(Base):
    __tablename__ = 'states'
    __table_args__ = (
        UniqueConstraint('state'),
    )
    id = Column(Integer, primary_key=True)
    state = Column(String(50))
    
class City(Base):
    __tablename__ = 'cities'
    __table_args__ = (
        CheckConstraint('LENGTH(city) > 0'),
    )
    id = Column(Integer, primary_key=True)
    city = Column(String(50))
    state_id = Column(Integer, ForeignKey('states.id'))

class Stadium(Base):
    __tablename__ = 'stadiums'
    __table_args__ = (
        CheckConstraint('LENGTH(stadium) > 0'),
    )
    id = Column(Integer, primary_key=True)
    stadium = Column(String(100))
    city_id = Column(Integer, ForeignKey('cities.id'))

class Team(Base):
    __tablename__ = 'teams'
    __table_args__ = (
        UniqueConstraint('team'),
    )
    id = Column(Integer, primary_key=True)
    team = Column(String(100))
    
class Superbowl(Base):
    __tablename__ = 'superbowls'
    id = Column(Integer, primary_key=True)
    date = Column(Date)
    sb = Column(String(50))
    winner_id = Column(Integer, ForeignKey('teams.id'))
    winner_points = Column(Integer)
    loser_id = Column(Integer, ForeignKey('teams.id'))
    loser_points = Column(Integer)
    mvp_id = Column(Integer, ForeignKey('players.id'))
    stadium_id = Column(Integer, ForeignKey('stadiums.id'))
```

```python
Base.metadata.create_all(engine)
```

```python
state_list.to_sql('states',engine, if_exists='append')
city_list.to_sql('cities',engine, if_exists='append')
stadium_list.to_sql('stadiums',engine, if_exists='append')
player_list.to_sql('players',engine, if_exists='append')
team_list.to_sql('teams',engine, if_exists='append')
superbowl_list.to_sql('superbowls',engine, if_exists='append')
```
